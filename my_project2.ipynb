```python
merged_df ['temp'] = merged_df['temp']. fillna(merged_df['temp'].mean())
merged_df ['atemp'] = merged_df['atemp']. fillna(merged_df['atemp'].mean())
merged_df ['hum'] = merged_df['hum']. fillna(merged_df['hum'].mean())
merged_df ['windspeed'] = merged_df['windspeed']. fillna(merged_df['windspeed'].mean())
```

```python
print("Mean of temp:", merged_df['temp'].mean())
print("Median of temp:", merged_df['temp'].median())
print("Mode of temp:", merged_df['temp'].mode()[0])
```
